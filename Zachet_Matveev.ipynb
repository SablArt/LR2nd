```python
import numpy as np
import matplotlib.pyplot as plt

N = int(input("Введите N: "))
x = float(input("Введите x: "))

S = 1
for i in range(1, N):
  S += x**(i)/np.math.factorial(i)

x_graph = np.arange(0,N,0.1)
y_graph = np.exp(x_graph)


plt.figure(1)
plt.plot(x,S, 'r*')
plt.plot(x_graph, y_graph)
print("сумма ", N, "чисел ряда равна", S)
```
